```python
#1. Find the Largest of Three Numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
if a >= b and a >= c:
 print("largest =", a)
elif b >= a and b >= c:
 print("largest =", b)
else:
 print("largest = ",c)
```

```python
#2. Check Whether a Year is a Leap Year
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
 print({year}," is a Leap Year")
else:
 print({year}," is NOT a Leap Year")
```

```python
#3. Check if a Character is a Vowel or Consonant
char = input("Enter a character: ")

if char in 'aeiou':
    print({char}," is a Vowel")
else:
    print({char} ,"is a Consonant")
```

```python
#4. Check if a Number is Divisible by Both 5 and 11
num = int(input("Enter a number: "))
if num % 5 == 0 and num % 11 == 0:
 print({num} ,"is divisible by both 5 and 11")
else:
 print({num}," is NOT divisible by both 5 and 11")
```

```python
#5. Sum of First N Natural Numbers (using while loop)
n = int(input("Enter a positive number N: "))
total = 0
i = 1
while i <= n:
 total += i
 i += 1
print("Sum of first {n} natural numbers = ",{total})
```

```python
#6. Multiplication Table (using while loop)
num = int(input("Enter a number: "))
print("Multiplication Table of:", {num})
print("-" * 20)
i = 1
while i <= 10:
 print(num ,"x" ,i,"=" ,{num * i})
 i += 1
```

```python
#7. Pyramid Pattern
rows = 5
i=1
while i < rows:
 print(" " * (rows - i), end="")
 print("*" * (2 * i - 1))
 i +=1
```

```python
#8. Number Triangle Pattern
i = 1
while i <= 5:
    j = 1
    while j <= i:
        print(j, end=" ")
        j += 1
    print()
    i += 1
```
